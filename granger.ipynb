```python
import numpy as np
import matplotlib.pyplot as plt
```

# Granger因果性

```python
t = np.arange(1, 101)
x = t/10 + np.random.normal(size=len(t))
y = x + np.random.normal(size=len(t))
```

```python
plt.plot(t, x)
plt.plot(t, y)
plt.show
```

```python
# x=>yの因果
data = np.array([[x],[y]]).reshape(100, 2)

from statsmodels.tsa.stattools import grangercausalitytests
maxlag = 1
results = grangercausalitytests(data, maxlag) # x: 2darray (data数, 2), x[:,0] => x[:,1]への因果を検定 
p_value = results[1]
print(p_value)
```

```python
# y=>xの因果
data = np.array([[y],[x]]).reshape(100, 2)

from statsmodels.tsa.stattools import grangercausalitytests
maxlag = 1
results = grangercausalitytests(data, maxlag) # x: 2darray (data数, 2), x[:,0] => x[:,1]への因果を検定 
p_value = results[1]
print(p_value)
```
